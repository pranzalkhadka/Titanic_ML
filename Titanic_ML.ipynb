```python
test = pd.read_csv('test.csv')
test.head(3)
```

```python
test["Age"] = test['Age'].fillna(test['Age'].median())
```

```python
test['Cabin'] = my_imputer.fit_transform(test[['Cabin']])
```

```python
test = test.drop(['Name','SibSp','Parch','Ticket','Fare','Embarked'],axis = 1)
```

```python
test.head(3)
```

```python
x_train = dataframe.drop("Survived",axis =1)
y_train = dataframe["Survived"]
```

```python
x_train
```

```python
y_train
```

```python
x_test = test.drop("PassengerId",axis = 1).copy()
```

```python
x_train["Sex"] = x_train["Sex"].map({'male':0,'female':1})
```

```python
#x_train = x_train.drop(['Cabin'],axis = 1)
x_train = x_train.drop(['PassengerId'],axis = 1)
```

```python
x_test["Sex"] = x_test["Sex"].map({'male':0,'female':1})
x_test = x_test.drop(['Cabin'],axis = 1)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
x_train.head(3)
```

```python
x_test.head(3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticregression = LogisticRegression()
logisticregression.fit(x_train,y_train)
```

```python
prediction = logisticregression.predict(x_test)
```

```python
accuracy = logisticregression.score(x_train,y_train)
```

```python
logistic_accuracy = accuracy * 100
print("The accuracy of logistic regression was found to be ",logistic_accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decisiontree = DecisionTreeClassifier()
decisiontree.fit(x_train,y_train)
```

```python
prediction = decisiontree.predict(x_test)
```

```python
Accuracy = decisiontree.score(x_train,y_train)
```

```python
dt_accuracy = Accuracy * 100
print("The accuracy by Decision Tree was found to be ",dt_accuracy)
```
