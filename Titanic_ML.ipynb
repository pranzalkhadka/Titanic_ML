```python
dataframe[['Sex','Pclass','Embarked','Survived','SibSp','Parch']] = dataframe[['Sex','Pclass','Embarked','Survived','SibSp','Parch']].astype("category")
```

```python
dataframe.info()
```

```python
dataframe["Age"] = dataframe['Age'].fillna(dataframe['Age'].median())
```

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer(strategy='most_frequent')
dataframe['Cabin'] = my_imputer.fit_transform(dataframe[['Cabin']])
```
